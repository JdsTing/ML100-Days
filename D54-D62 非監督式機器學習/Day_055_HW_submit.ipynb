### 用 iris (dataset.load_iris()) 資料嘗試跑 kmeans (可以測試不同的群數 , init 等)

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

### 載入 相關套件 並 執行 kmean 實驗 ...

- 實驗測試不同的群數
- 實驗測試不同的初始值
- 呈現結果

```python
def plot_cluster(X, y, labels):
    fig = plt.figure(figsize=(8, 6))
    ax = fig.gca(projection='3d')
    ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap=plt.cm.jet)
    ax.set_xlabel(labels[0])
    ax.set_ylabel(labels[1])
    ax.set_zlabel(labels[2])
    ax.view_init(elev=30, azim=225)

# original data labels
plot_cluster(X, y, iris.feature_names[1:])
```

```python
plt.scatter(X[:,0],X[:,1],marker='o',edgecolor='black',c=y)
plt.grid()
plt.show()
```

```python
# 2 clusters

estimator = KMeans(n_clusters=2)
estimator.fit(X)
y_pred = estimator.labels_

plot_cluster(X, y_pred, iris.feature_names[1:])
```

```python
# 3 clusters

estimator = KMeans(n_clusters=3)
estimator.fit(X)
y_pred = estimator.labels_

plot_cluster(X, y_pred, iris.feature_names[1:])
```

```python
#分三群
km = KMeans(n_clusters=3,init='random',n_init=10,max_iter=300,tol=1e-04,random_state=0)
y_km = km.fit_predict(X)
plt.scatter(X[:,0],X[:,1],marker='o',edgecolor='black',c=y_km)
plt.grid()
plt.show()
print(X)
```

```python
# 4 clusters

estimator = KMeans(n_clusters=4)
estimator.fit(X)
y_pred = estimator.labels_

plot_cluster(X, y_pred, iris.feature_names[1:])
```

```python
#計算不同分組數量的集群內誤差平方和，並劃出圖表
distortions = []
for i in range(1,11):
    km = KMeans(n_clusters=i,init='random',n_init=10,max_iter=100,tol=1e-04,random_state=0)
    km.fit(X)
    distortions.append(km.inertia_)

plt.plot(range(1,11),distortions,marker= 'o')
plt.xlabel('Number of Clusters')
plt.ylabel('Distortion')
plt.show()
```

圖表顯示，在群數為三的時候就出現轉折，超過三組的話，分群效果就不太明顯了。因此iris數據集的最佳分群參數是 ３
